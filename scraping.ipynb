```python
import requests
r=requests.get("https://www.imdb.com/chart/top/")
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(r.text)#lister-list
```

```python
results=soup.find('tbody', {'class': 'lister-list'}).findAll('tr')
```

```python
#extract (imageSource , movieImdbLink , movieRank,movieName, movieYear ,imdbRating)
first_row=results[0]
```

```python
imageSource=first_row.contents[1].find('a').find('img')['src']
movieRank=first_row.contents[3].text.split('\n')[1].strip()[0:-1]
movieName=first_row.contents[3].text.split('\n')[2].strip()
movieYear=first_row.contents[3].find('span').text.strip('(').strip(')')
imdbRating=first_row.contents[5].find('strong').text
movieImdbLink=first_row.findAll('a')[1]['href']
[movieName,movieRank,imdbRating,movieYear,imageSource,movieImdbLink]
```

```python
len(results)
```

```python
#save whole movies in a records in order to save them in a dataSet
records=[]
for i in range(len(results)):
    itr_row=results[i]
    imageSource=itr_row.contents[1].find('a').find('img')['src']
    movieRank=itr_row.contents[3].text.split('\n')[1].strip()[0:-1]
    movieName=itr_row.contents[3].text.split('\n')[2].strip()
    movieYear=itr_row.contents[3].find('span').text.strip('(').strip(')')
    imdbRating=itr_row.contents[5].find('strong').text
    movieImdbLink=itr_row.findAll('a')[1]['href']
    records.append((movieRank,movieYear,imdbRating,movieName,imageSource,movieImdbLink))
```

```python
records[0:3]
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['rank','year','rate','movie_name','image','link'])
df
```

```python
df.to_csv('topImdb.csv',index=False,encoding='utf-8')
df=pd.read_csv('topImdb.csv',encoding='utf-8')
```
